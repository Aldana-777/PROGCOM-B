#Descripción Florería Aldana en Turtle


```python
import turtle

# Función inicial para el funcionamiento
screen = turtle.Screen()
screen.title("Casa desde 5 perspectivas con colores personalizados")
screen.bgcolor("lightgreen")  # Césped
t = turtle.Turtle()
t.speed(0)

# Función para dibujar el primer rectángulo
def dibujar_rectangulo(t, ancho, alto, color):
    t.fillcolor(color)
    t.begin_fill()
    for _ in range(2):
        t.forward(ancho)
        t.left(90)
        t.forward(alto)
        t.left(90)
    t.end_fill()

# Función para dibujar triángulo para el techo
def dibujar_triangulo(t, base, altura, color):
    t.fillcolor(color)
    t.begin_fill()
    t.forward(base)
    t.left(135)
    t.forward(altura)
    t.left(90)
    t.forward(altura)
    t.left(135)
    t.end_fill()

# Función para dibujar la puerta
def dibujar_puerta(t, ancho, alto, color):
    dibujar_rectangulo(t, ancho, alto, color)

# Función para dibujar ventana
def dibujar_ventana(t, lado, color):
    dibujar_rectangulo(t, lado, lado, color)

# Función de la Vista frontal de Floristería Aldana
def casa_frontal(x, y):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pendown()
    dibujar_rectangulo(t, 100, 100, "limegreen")

    t.penup()
    t.goto(x, y + 100)
    t.setheading(0)
    t.pendown()
    dibujar_triangulo(t, 100, 70, "black")

    # Función para la Puerta
    t.penup()
    t.goto(x + 35, y)
    t.setheading(0)
    t.pendown()
    dibujar_puerta(t, 30, 50, "deepskyblue")
    ventana_x = x + 40
    ventana_y = y + 60

    # Función de la flor delantera
    t.penup()
    t.goto(ventana_x + 10, ventana_y + 10 - 30)
    t.setheading(0)
    t.pendown()
    t.fillcolor("black")
    t.begin_fill()
    t.circle(30)
    t.end_fill()

    # Función de los pétalos amarillos de la flor
    petalo_radio = 10
    centro_x = ventana_x + 10
    centro_y = ventana_y + 10
    t.fillcolor("yellow")
    for angle in range(0, 360, 45):  # 8 pétalos
        t.penup()
        t.goto(centro_x, centro_y)
        t.setheading(angle)
        t.forward(30)  # distancia desde el centro
        t.pendown()
        t.begin_fill()
        t.circle(petalo_radio)
        t.end_fill()

    # Función de la ventana blanca encima del círculo
    t.penup()
    t.goto(ventana_x, ventana_y)
    t.setheading(0)
    t.pendown()
    dibujar_ventana(t, 20, "white")

    # Función de la ventana blanca encima del círculo
    t.penup()
    t.goto(ventana_x, ventana_y)
    t.setheading(0)
    t.pendown()
    dibujar_ventana(t, 20, "white")

# Función de la vista trasera de la Florería Aldana
def casa_trasera(x, y):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pendown()
    dibujar_rectangulo(t, 100, 100, "limegreen")
    t.penup()
    t.goto(x, y + 100)
    t.setheading(0)
    t.pendown()
    dibujar_triangulo(t, 100, 70, "black")

    # Triángulo debajo del techo
    t.penup()
    t.goto(x + 15, y + 100)
    t.setheading(0)
    t.pendown()
    dibujar_triangulo(t, 70, 49, "gold")

    # Círculo blanco en techo
    t.penup()
    t.goto(x + 50, y + 100)
    t.setheading(0)
    t.pendown()
    t.fillcolor("white")
    t.begin_fill()
    t.circle(10)
    t.end_fill()

       # Nube
    nube_x = x + 17  # 3 unidades a la izquierda
    nube_y = y + 43  # subida total de 8 unidades desde el original
    t.fillcolor("gray")
    t.pencolor("gray")

    for dx, dy in [
        (0, 0), (18, 8), (36, 0), (54, 8), (72, 0),
        (18, -12), (36, -12), (54, -12)
    ]:
        t.penup()
        t.goto(nube_x + dx, nube_y + dy)
        t.setheading(0)
        t.pendown()
        t.begin_fill()
        t.circle(12)
        t.end_fill()

    # Ventana
    ventana_ancho = 60
    ventana_alto = 30
    ventana_x = x + (100 - ventana_ancho) / 2
    ventana_y = y + 48  # bajada total de 12 unidades desde el original

    t.penup()
    t.goto(ventana_x, ventana_y)
    t.setheading(0)
    t.pendown()
    t.fillcolor("white")
    t.begin_fill()
    for _ in range(2):
        t.forward(ventana_ancho)
        t.left(90)
        t.forward(ventana_alto)
        t.left(90)
    t.end_fill()
    t.pencolor("black")
    t.pensize(1)
    t.penup()
    t.goto(ventana_x + ventana_ancho / 3, ventana_y)
    t.setheading(90)
    t.pendown()
    t.forward(ventana_alto)
    t.penup()
    t.goto(ventana_x + 2 * ventana_ancho / 3, ventana_y)
    t.setheading(90)
    t.pendown()
    t.forward(ventana_alto)

    # Riachuelo
    t.penup()
    t.goto(x, y + 35)
    t.setheading(-25)
    t.pendown()
    t.pensize(12)
    t.pencolor("skyblue")
    t.forward(100)
    t.pensize(1)
    t.pencolor("black")

    # Función del lado izuierdo de la Floristería Aldana
def casa_lado_izquierdo(x, y):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pendown()
    dibujar_rectangulo(t, 60, 100, "limegreen")
    t.penup()
    t.goto(x, y + 100)
    t.setheading(0)
    t.pendown()
    dibujar_triangulo(t, 60, 42, "black")
    t.penup()
    t.goto(x + 20, y + 60)
    # Tres palos verticales en la pared izquierda
    t.pencolor("lightgray")
    t.pensize(4)
    for offset in [10, 25, 40]:
        t.penup()
        t.goto(x + offset, y)
        t.setheading(90)
        t.pendown()
        t.forward(60)
    t.pensize(1)
    t.pencolor("black")

# Función del lado derecho de la Floristería Aldana
def casa_lado_derecho(x, y):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pendown()
    dibujar_rectangulo(t, 60, 100, "limegreen")
    t.penup()
    t.goto(x, y + 100)
    t.setheading(0)
    t.pendown()
    dibujar_triangulo(t, 60, 42, "black")
    t.penup()
    t.goto(x + 20, y + 60)

# Función vista desde arriba (techo) de la floristería Aldana
def casa_desde_arriba(x, y):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pendown()
    t.fillcolor("black")
    t.begin_fill()
    for _ in range(2):
        t.forward(100)
        t.left(90)
        t.forward(80)
        t.left(90)
    t.end_fill()

    # Líneas de tejas
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pencolor("gray")
    for _ in range(10):
        t.forward(10)
        t.pendown()
        t.left(90)
        t.forward(80)
        t.backward(80)
        t.right(90)
        t.penup()

# Dibujo General de La descripción de la Florería Aldana
casa_frontal(-250, 100)
casa_trasera(-80, 100)
casa_lado_izquierdo(100, 100)
casa_lado_derecho(220, 100)
casa_desde_arriba(-50, -150)

t.hideturtle()
screen.mainloop()
```
