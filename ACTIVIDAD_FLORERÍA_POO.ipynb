FUNCIÓN FLORERÍA

```python
import turtle

# Configuración inicial
screen = turtle.Screen()
screen.title("Casa desde 5 perspectivas con colores personalizados")
screen.bgcolor("lightgreen")  # Césped
t = turtle.Turtle()
t.speed(0)

# Función para dibujar un rectángulo
def dibujar_rectangulo(t, ancho, alto, color):
    t.fillcolor(color)
    t.begin_fill()
    for _ in range(2):
        t.forward(ancho)
        t.left(90)
        t.forward(alto)
        t.left(90)
    t.end_fill()

# Función para dibujar un triángulo (techo)
def dibujar_triangulo(t, base, altura, color):
    t.fillcolor(color)
    t.begin_fill()
    t.forward(base)
    t.left(135)
    t.forward(altura)
    t.left(90)
    t.forward(altura)
    t.left(135)
    t.end_fill()

# Función para dibujar una puerta
def dibujar_puerta(t, ancho, alto, color):
    dibujar_rectangulo(t, ancho, alto, color)

# Función para dibujar una ventana
def dibujar_ventana(t, lado, color):
    dibujar_rectangulo(t, lado, lado, color)

# Vista frontal
def casa_frontal(x, y):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pendown()
    dibujar_rectangulo(t, 100, 100, "limegreen")
    t.penup()
    t.goto(x, y + 100)
    t.setheading(0)
    t.pendown()
    dibujar_triangulo(t, 100, 70, "black")
    t.penup()
    t.goto(x + 35, y)
    t.setheading(0)
    t.pendown()
    dibujar_puerta(t, 30, 50, "deepskyblue")
    t.penup()
    t.goto(x + 10, y + 60)
    t.pendown()
    dibujar_ventana(t, 20, "white")
    t.penup()
    t.goto(x + 70, y + 60)
    t.pendown()
    dibujar_ventana(t, 20, "white")

# Vista trasera
def casa_trasera(x, y):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pendown()
    dibujar_rectangulo(t, 100, 100, "limegreen")
    t.penup()
    t.goto(x, y + 100)
    t.setheading(0)
    t.pendown()
    dibujar_triangulo(t, 100, 70, "black")
    t.penup()
    t.goto(x + 40, y + 40)
    t.pendown()
    dibujar_ventana(t, 20, "white")

# Vista lateral izquierda
def casa_lado_izquierdo(x, y):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pendown()
    dibujar_rectangulo(t, 60, 100, "limegreen")
    t.penup()
    t.goto(x, y + 100)
    t.setheading(0)
    t.pendown()
    dibujar_triangulo(t, 60, 42, "black")
    t.penup()
    t.goto(x + 20, y + 60)
    t.pendown()
    dibujar_ventana(t, 20, "white")

# Vista lateral derecha
def casa_lado_derecho(x, y):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pendown()
    dibujar_rectangulo(t, 60, 100, "limegreen")
    t.penup()
    t.goto(x, y + 100)
    t.setheading(0)
    t.pendown()
    dibujar_triangulo(t, 60, 42, "black")
    t.penup()
    t.goto(x + 20, y + 60)
    t.pendown()
    dibujar_ventana(t, 20, "white")

# Vista desde arriba (techo)
def casa_desde_arriba(x, y):
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pendown()
    t.fillcolor("black")
    t.begin_fill()
    for _ in range(2):
        t.forward(100)
        t.left(90)
        t.forward(80)
        t.left(90)
    t.end_fill()

    # Líneas de tejas
    t.penup()
    t.goto(x, y)
    t.setheading(0)
    t.pencolor("gray")
    for _ in range(10):
        t.forward(10)
        t.pendown()
        t.left(90)
        t.forward(80)
        t.backward(80)
        t.right(90)
        t.penup()

# Dibujo de las 5 vistas
casa_frontal(-250, 100)
casa_trasera(-80, 100)
casa_lado_izquierdo(100, 100)
casa_lado_derecho(220, 100)
casa_desde_arriba(-50, -150)

t.hideturtle()
screen.mainloop()
```
