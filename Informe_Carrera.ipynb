```python
# 游늵 C치lculo y gr치fico del ahorro acumulado en 10 a침os

# Par치metros iniciales
salario_mensual = 1_200_000       # Salario mensual inicial
reajuste_anual = 0.05             # Reajuste salarial anual (5%)
interes_anual = 0.24              # Inter칠s compuesto anual (24%)
a침os = 10                         # N칰mero de a침os

# Inicializaci칩n
ahorro_acumulado = 0
ahorros_por_a침o = []

# C치lculo del ahorro acumulado a침o a a침o
for a침o in range(1, a침os + 1):
    salario_diciembre = salario_mensual * ((1 + reajuste_anual) ** (a침o - 1))
    ahorro_acumulado = ahorro_acumulado * (1 + interes_anual) + salario_diciembre
    ahorros_por_a침o.append(round(ahorro_acumulado))

# Mostrar resultado final
print("Ahorro acumulado al final de 10 a침os: ${:,.0f}".format(ahorro_acumulado))

# Mostrar evoluci칩n a침o a a침o
for i in range(a침os):
    print("A침o {}: ${:,.0f}".format(i + 1, ahorros_por_a침o[i]))

# Gr치fico (requiere matplotlib, ya viene preinstalado en Colab)
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(range(1, a침os + 1), ahorros_por_a침o, marker='o', color='blue', linewidth=2)
plt.title('Evoluci칩n del Ahorro Acumulado en 10 A침os', fontsize=14)
plt.xlabel('A침o', fontsize=12)
plt.ylabel('Ahorro Acumulado ($)', fontsize=12)
plt.grid(True)
plt.xticks(range(1, a침os + 1))
plt.tight_layout()
plt.show()

```
