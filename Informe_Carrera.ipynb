```python
# 📊 Cálculo y gráfico del ahorro acumulado en 10 años

# Parámetros iniciales
salario_mensual = 1_200_000       # Salario mensual inicial
reajuste_anual = 0.05             # Reajuste salarial anual (5%)
interes_anual = 0.24              # Interés compuesto anual (24%)
años = 10                         # Número de años

# Inicialización
ahorro_acumulado = 0
ahorros_por_año = []

# Cálculo del ahorro acumulado año a año
for año in range(1, años + 1):
    salario_diciembre = salario_mensual * ((1 + reajuste_anual) ** (año - 1))
    ahorro_acumulado = ahorro_acumulado * (1 + interes_anual) + salario_diciembre
    ahorros_por_año.append(round(ahorro_acumulado))

# Mostrar resultado final
print("Ahorro acumulado al final de 10 años: ${:,.0f}".format(ahorro_acumulado))

# Mostrar evolución año a año
for i in range(años):
    print("Año {}: ${:,.0f}".format(i + 1, ahorros_por_año[i]))

# Gráfico (requiere matplotlib, ya viene preinstalado en Colab)
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(range(1, años + 1), ahorros_por_año, marker='o', color='blue', linewidth=2)
plt.title('Evolución del Ahorro Acumulado en 10 Años', fontsize=14)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Ahorro Acumulado ($)', fontsize=12)
plt.grid(True)
plt.xticks(range(1, años + 1))
plt.tight_layout()
plt.show()

```
